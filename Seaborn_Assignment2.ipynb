# Q1: Import seaborn library

```python
# write your code here ^_^
import seaborn as sns
```

```python
tips_dataset = sns.load_dataset('tips')
```

```python
tips_dataset.head()
```

# Q2: Create a histogram, where the x-axis represents the tip column
- Change the style to 'whitegrid'.
- Split the data out by the day column's categories. Hint: use hue argument.
- Handle the overlapping columns by switching the element argument over to step.

```python
# write your code here ^_^

# Set the style to 'whitegrid'
sns.set_style('whitegrid')
plt.figure(figsize=(15,8))

# Create the histogram with split categories
sns.histplot(data=tips_dataset, x='tip', hue='day', element='step')

# Set the plot title and labels
plt.title('Tip Distribution by Day')
plt.xlabel('Tip Amount')
plt.ylabel('Count')

# Display the plot
plt.show()
```

## Q2.1 Write your top three findings


- Higher Tip Amounts on Weekends: On weekends, especially on Friday and Saturday nights, there are significantly higher tip amounts. This suggests that these are the nights when restaurants experience the most profitable revenue.

-  Lower Tip Amounts on Weekdays: In contrast, the average tip amount is much lower on weekdays. This may indicate that customers during the week may be less inclined to tip due to time constraints or financial limitations.

- Uneven Tip Distribution by Day: The tip distribution varies significantly across the week. While Thursday and Sunday evenings have the highest average tip amounts, Wednesday evenings have the lowest average tip amounts. This may indicate that certain days of the week tend to be more profitable for restaurants, or that certain demographics may be more generous with their tips on certain days.


# Q3: [Optional] Create a displot, where the x-axis represents the day column
- Split the data out by the sex column's categories. Hint: use hue argument.

- Shrink the bars down using shrink argument to have a nice gap between each of them.

```python
# write your code here ^_^

# Set the hue order to ['Female', 'Male']
hue_order = ['Female', 'Male']

# Set the style to 'whitegrid'
sns.set(style='whitegrid')

# Create the displot
sns.displot(data=tips_dataset, x='day', hue='sex', multiple='dodge', shrink=0.8, palette='Set1', hue_order=hue_order)

# Display the plot
plt.show()
```

# Q4: Create a ecdf plot, where the x-axis represents the tip column
- Change the style to 'darkgrid'.
- Change the font scale to 1.25.
- Split the data out by the time column's categories. Hint: use hue argument.
- Change the lines color to 'summer'.
- Change the line width to 3.

```python
# write your code here ^_^

# Set the style to 'darkgrid'
sns.set(style='darkgrid')

# Set the font scale to 1.25
sns.set(font_scale=1.25)

# Split the data by the 'time' column's categories and useing the 'hue' argument
sns.ecdfplot(data=tips_dataset, x='tip', hue='time', palette='summer', linewidth=3)

# Display the plot
plt.show()
```

## Q4.1 Write your top three findings
- Peak consumption hours: Throughout the day, the proportion of lunches, dinners, and other meals significantly peaks at certain hours. This is likely due to the work schedule, with most people consuming meals during their breaks or at the end of the day.

-  Relatively lower consumption during late-night hours: On the other hand, the consumption of tips and other meals declines significantly as the day progresses.This may indicate that many individuals prefer to consume meals and snacks earlier in the day.

- Gender and age-related variations: While the general pattern of meal consumption throughout the day remains consistent, there are subtle differences in the proportion of tips and other meals consumed based on gender and age. For example, women tend to consume more meals during the evening, while men tend to consume more during the morning. Similarly, older individuals, on average, consume fewer meals throughout the day compared to younger individuals. These differences could be attributed to factors such as individual preferences, dietary habits, and health conditions.

# Q5: [Optional] Create a box plot, where the x-axis represents the day column and the y-axis represents the tip column
- Change the style to 'whitegrid'.
- Split the data out by the sex column's categories. Hint: use hue argument.
- Change the bars color to 'Paired'.
- Reset the hue order to ['Female', 'Male'].
- Change the line width to 2.5.

```python
# write your code here ^_^


# Set the hue order to ['Female', 'Male']
hue_order = ['Female', 'Male']

# Set the style to 'whitegrid'
sns.set(style='whitegrid')

# Create the box plot
sns.boxplot(data=tips_dataset, x='day', y='tip', hue='sex', palette='Paired', hue_order=hue_order, linewidth=2.5)

# Display the plot
plt.show()
```

## Q5.1 Write your top three findings
- Female and male tips were higher on Thursdays.

- Male and female tips were significantly lower on Saturdays compared to other days.

- Outliers are also found on Thursdays, Fridays, and Saturdays more than on any other day.

- Also on Saturdays, there are outliers in male-only.

- Overall, male tips were slightly higher than female tips across all days.

# Q6: Create a violin plot, where the x-axis represents the time column and the y-axis represents the tip column
- Split the data out by the sex column's categories. Hint: use hue argument.
- Split the violin so that each violin contains both sex column's categories. Hint: use split argument.
- Change the shape of the inner part of the violin to 'quartiles'.
- Change the violin color to 'Pastel1'.
- Reset the hue order to ['Female', 'Male'].
- Change the line width to 2.
- Change the legend's location to The upper left.

```python
# write your code here ^_^

# Set the hue order to ['Female', 'Male']
hue_order = ['Female', 'Male']

# Set the style to 'whitegrid'
sns.set(style='whitegrid')

# Create the violin plot
sns.violinplot(data=tips_dataset, x='time', y='tip', hue='sex', split=True, inner='quartiles', palette='Pastel1', hue_order=hue_order, linewidth=2)

# Set the legend's location to upper left
plt.legend(loc='upper left')

# Display the plot
plt.show()
```

## Q6.1 Write your top three findings
-Lunch: Tips are higher during dinner compared to lunch.

-shows that the total bill for men is higher than the total bill for women on an average.

-Meal Times: Overall, tips increase throughout the day. Employees are more likely to receive higher tips during peak times when businesses are busier and potentially generating more revenue.

# Q8: Create a correlation heatmap
- Adjust the figure's size to be (8,5).
- Change the color map to 'RdBu'.
- Add annotations to the heatmap.
- Set the minimum value to be -1.
- Set the maximum value to be 1.

```python
# write your code here ^_^

# Calculate the correlation matrix
corr_matrix = tips_dataset.corr()

# Set figure size
plt.figure(figsize=(8, 5))

# Create the correlation heatmap
sns.heatmap(corr_matrix, cmap='RdBu', annot=True, vmin=-1, vmax=1)

# Set plot title
plt.title('Correlation Heatmap')

# Show the plot
plt.show()
```
